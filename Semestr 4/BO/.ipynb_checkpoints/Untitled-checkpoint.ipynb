```r
calc.lambda <- function(m, L, f) {
    l <- c()
    for (i in 0:(m+L)) {
        if (i <= m) {
            l <- c(l, f)
        } else {
            l <- c(l, 0)
        }
    }
    return(l)
}

calc.mi <- function(m, L, f) {
    mi <- c()
    for (i in 0:(m+L)) {
        mi <- c(mi, f*min(i,m))
    }
    return(mi)
}

calc.p <- function(lambda, mi) {
    p0 <- 1
    tmp <- 1
    for (i in 2:length(mi)) {
        tmp <- tmp * lambda[i-1] / mi[i]
        p0 <- p0 + tmp
    }
    p0 <- 1 / p0
    p <- c(p0)
    tmp <- 1
    for (i in 2:length(mi)) {
        tmp <- tmp * lambda[i-1] / mi[i]
        p <- c(p, p0 * tmp)
    }
    return(p)
}

calc.N.avg <- function(p) {
    N <- 0
    for (i in 1:length(p)) {
        N <- N + (i-1) * p[i]
    }
    return(N)
}

calc.lambda.avg <- function(lambda, p) {
    lambda.avg <- 0
    for (i in 1:length(p)) {
        lambda.avg <- lambda.avg + lambda[i] * p[i]
    }
    return(lambda.avg)
}

precision<-5

m <- 2 #liczba stanowisk
L <- 1 #pojemność kolejkk
f.in <- 8 #ile przybywa
f.out <- 12 #wydajność
lambda <- calc.lambda(m,L,f.in) #myjnie
# lambda <- c(8,4,0) #końcówki (2*l, l ,0)
mi <- calc.mi(m,L,f.out) #myjnie
# mi <- c(0,10,10) #końcówki
p <- round(calc.p(lambda, mi), precision)

N.avg <- round(calc.N.avg(p), precision)

lambda.avg <- calc.lambda.avg(lambda, p)

T.avg <- N.avg / lambda.avg

W <- p[length(p)] / lambda.avg

cat("p          = [", p, "]\n")
cat("N.avg      =  ", N.avg, "\n")
cat("lambda.avg =  ", lambda.avg, "\n")
cat("T.avg      =  ", round(T.avg, precision), "\n")
cat("W          =  ", round(W, precision), "\n")
```
