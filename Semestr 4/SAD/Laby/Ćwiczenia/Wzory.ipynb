```r
install.packages("moments")
library(moments)
moment_centralny <- function(dane, rzad){
    mean((dane - mean(dane)) ^ rzad)
}

moment_centralny_rounded <- function(dane,rzad){
    round(mean((dane - mean(dane)) ^ rzad),2)
}

wariancja <- function(dane){
    sum((dane-mean(dane)) ^ 2) / (length(dane)-1)
}

wariancja_populacji <- function(dane){
    sum((dane-mean(dane)) ^ 2) / (length(dane))
}

odchylenie_std <- function(dane){
    sqrt(wariancja(dane))
}

odchylenie_std_populacji <- function(dane){
    sqrt(wariancja_populacji(dane))
}

skosnosc <- function(dane){    #skośność
    moment_centralny(dane,3) / (odchylenie_std_populacji(dane)^3)
}

kurtoza <- function(dane){    #kurtoza
    moment_centralny(dane,4) / odchylenie_std(dane)^4
}

kurtoza_min3 <- function(dane){    #kurtoza - 3
    moment_centralny(dane,4) / odchylenie_std(dane)^4 - 3
}

mode <- function(dane){   #dominanta
    uniqv <- unique(dane)
    uniqv[which.max(tabulate(match(dane,uniqv)))]
}

alt_skewness <- function(dane){    #alternatywna skośexpected valueność
    (mean(dane)-mode(dane)) / odchylenie_std(dane)
}


#Lab 4

gestosc <- function(x, mean=0, sd=1){    #gęstość rozkładu normalnego
    1/(sd*sqrt(2*pi)) * exp(-0.5*((x-mean)/sd)^2)
}

#cumsum(x)    #sumuje wszystkie elementy od 1-ego do i-tego

delta = 0.01
z <- -2000:2000*delta

dystrybuanta <- cumsum(delta * dnorm(z, 0, 1))


#suma odchyleń zawsze wynosi 0
# średnia x   +n
# wariancja y nie zmienia się
# odch_std z  nie zmienia się
```
