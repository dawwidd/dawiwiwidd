```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

1) Log file is loaded

```python
file = open("log.txt", "r") 
lines = file.readlines()
for l in range(3): print(lines[l][:-1])
```

2) Cass "Request" parses and keeps data of a single requests. Complete the TODO, where a given "line" (string) should be parsed in order to dervie the ip of the host making the request, time of the request (in seconds), and target page (name).

```python
class Request:
    def __init__(self, line):
        # TODO DO PARSING HERE
        splittedLog = line.split()
        
        self.ip = splittedLog[0]
        # convert hours + minutes + seconds -> total number of seconds
        self.timeUnformatted = splittedLog[1]
        self.time = int(self.timeUnformatted[-2:]) + int(self.timeUnformatted[-5:-3]) * 60 + int(self.timeUnformatted[-8:-6]) * 60 * 60
        self.page = splittedLog[4]
        
    def __repr__(self):
        return self.ip + "     " +  str(datetime.timedelta(seconds=self.time)) + " -> " + self.page
```

3) Each line of log.txt is parsed and the Request objects are added to the list

```python
reqs = [Request(line) for line in lines]
for l in range(10): print(reqs[l]) 
```

# Exercise

Your task is to analyze the log file. You should indentify the users and sessions. Then, compute the following: <br>
3.1) the number of sessions (should be 1000 :) ), <br>
3.2) the number of users, <br>
3.3) the average number of sessions per one user <br>
3.4) use the distribution plot to visualize the sessions' lengths (minutes);  <br>
3.5) use the distribution plot to visualize the sessions' lengths (the number of requests) <br>
3.6) use the distribution plot to visualize sessions' starting times (hours). <br>
3.7) which web page is the most common entry point?; for each page plot the number of times it has been an entry page,  <br>
3.8) which web page is the most common departure point?; for each page plot the number of times the user stopped a session while being on this page, <br>
3.9) for each page plot the average time the user stayed on this page <br>
3.10) print the fist 10 sessions (paths). <br>

You may find the following classes helpful and data.

```python
class Session:
    def __init__(self, request):
        self.startTime = request.time
        self.stopTime = request.time
        self.requests = []
        self.requests.append(request)
        self.isActive = True
        
    def __repr__(self):
        print("SESSION")
        print(self.startTime)
        print(self.stopTime)
        STR = ""
        for r in self.requests:
            print(r)
            STR += ("  " + r.page + "    " + str(datetime.timedelta(seconds=r.time)) + "\n")
        return STR
    
    def closeSession(self):
        self.isActive = False
        
    def addRequest(self, request):
        self.stopTime = request.time
        self.requests.append(request)
        
    def getSessionLengthInMinutes(self):
        lengthInSeconds = self.stopTime - self.startTime
        lengthInMinutes = lengthInSeconds / 60
        return lengthInMinutes
    
    def getSessionStartingHour(self):
        startHour = self.startTime / 3600
        return startHour
    
    def timeSpentOnPageInMinutes(self, request):
        nextRequestIndex = self.requests.index(request) + 1
        timeSpent = self.requests[nextRequestIndex].time - request.time
        return timeSpent

class Computer:
    def __init__(self, ip, session):
        self.ip = ip
        self.sessions = []
        self.sessions.append(session)
        
    def __repr__(self):
        return self.ip
    
    def addSession(self, session):
        self.sessions.append(session)

PAGES = ["Shop.html", "AddProduct.html", "Summary.html", "Payment.html","About.html", "Contact.html"]

LINKS = [[1, 4],
         [2,4],
         [3],
         [],
         [1,5],
         []]

MAX_SESSION_TIME = 34 * 60

def shouldAppendRequestToSession(firstRequest, lastRequest, currentRequest):
    outgoingPagesCountForRequest = len(LINKS[PAGES.index(lastRequest.page)])
    if currentRequest.time < lastRequest.time or currentRequest.time - MAX_SESSION_TIME > firstRequest.time or outgoingPagesCountForRequest == 0:
        return False
    else:
        return True

users = []
computerIPs = []
for req in reqs:
    if req.ip not in computerIPs:
        computerIPs.append(req.ip)
        users.append(Computer(req.ip, Session(req)))
    for i, user in enumerate(users):
        if req.ip != user.ip:
            continue
        for session in user.sessions:
            firstSessionRequest = session.requests[0]
            lastSessionRequest = session.requests[-1]
            requestAppendedToSession = shouldAppendRequestToSession(firstSessionRequest, lastSessionRequest, req)
            if not session.isActive:
                continue
            elif requestAppendedToSession:
                session.addRequest(req)
            else:
                session.closeSession()
                user.addSession(Session(req))
                break
        

print("3.1 Number of users:", len(users))

allSessions = []
for user in users:
    for session in user.sessions:
        allSessions.append(session)
        
print("3.2 Number of all sessions: ", len(allSessions))

print("3.3 Average number of sessions per user:", len(allSessions)/len(users))

print("3.4 Length of sessions distribution")
allSessionsLengthsInMinutes = []
for session in allSessions:
    allSessionsLengthsInMinutes.append(session.getSessionLengthInMinutes())

plt.hist(allSessionsLengthsInMinutes, 25, alpha=0.5)
plt.show()

print("3.5 Number of requests per session distribution")
allSessionsNumberOfRequests = []
for session in allSessions:
    allSessionsNumberOfRequests.append(len(session.requests))
plt.hist(allSessionsNumberOfRequests, 25, alpha=0.5)
plt.show()

print("3.6 Starting hours of sessions distribution")
allSessionsStartingHours = []
for session in allSessions:
    allSessionsStartingHours.append(session.getSessionStartingHour())
plt.hist(allSessionsStartingHours, 25, alpha=0.5)
plt.show()


sessionsStartingPagesCounts = np.zeros(len(PAGES))
sessionsDeparturePagesCounts = np.zeros(len(PAGES))
for session in allSessions:
    sessionsStartingPagesCounts[PAGES.index(session.requests[0].page)] += 1
    sessionsDeparturePagesCounts[PAGES.index(session.requests[-1].page)] += 1
    
print("3.7 Counts of starting pages")
plt.bar(PAGES, sessionsStartingPagesCounts, align='center', alpha=0.5)
plt.xticks(rotation=45)
plt.show()

print("3.8 Counts of departure pages")
plt.bar(PAGES, sessionsDeparturePagesCounts, align='center', alpha=0.5)
plt.xticks(rotation=45)
plt.show()


print("3.9 Average time spent on each page")
pagesTimesSpent = np.zeros(len(PAGES))
pagesVisits = np.zeros(len(PAGES))
for session in allSessions:
    for request in session.requests[:-1]:
        pagesTimesSpent[PAGES.index(request.page)] += session.timeSpentOnPageInMinutes(request)
        pagesVisits[PAGES.index(request.page)] += 1
        
    request = session.requests[-1]
    pagesTimesSpent[PAGES.index(request.page)] += MAX_SESSION_TIME + session.startTime - session.stopTime
    pagesVisits[PAGES.index(request.page)] += 1
    
averageTimesOnPages = (pagesTimesSpent / pagesVisits) / 60

plt.bar(PAGES, averageTimesOnPages, align='center', alpha=0.5)
plt.xticks(rotation=45)
plt.show()                                                      


print("3.10 First ten sessions:")

for i, session in enumerate(allSessions[:10]):
    print("Session", i+1, end=" ")
    for i, request in enumerate(session.requests):
        if request == session.requests[-1]:
            print(request.page)
            break
        else:
            print(request.page, ", ", end="")
```
